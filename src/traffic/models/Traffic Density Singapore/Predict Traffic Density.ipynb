## Train YOLOv8 CLS Model

```python
from pathlib import Path

DATA_PATH = data_path = (
    Path.cwd().parent.parent.parent.parent  # Repository root
    / "data"
    / "traffic"
    / "singapore-traffic-density"
)

MODEL_PATH = 'runs/classify/train/weights/best.pt'
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n-cls.pt')
model.train(data=DATA_PATH, epochs=10, imgsz=640)
```

```python
best_model = YOLO(MODEL_PATH)

metrics = best_model.val()
```

```python
results = best_model('https://media.wired.com/photos/593256b42a990b06268a9e21/master/w_2560%2Cc_limit/traffic-jam-getty.jpg')
print(best_model.names[results[0].probs.top1])
```
