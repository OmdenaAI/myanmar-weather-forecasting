```python
import cv2
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = (
    Path.cwd().parent.parent.parent  # Repository root
    / "data"
    / "traffic"
    / "car-object-detection"
)

train_img_folder = os.path.join(data_path, 'training_images')
bbox_csv = os.path.join(data_path, 'train_solution_bounding_boxes (1).csv')
```

```python
print('Number of Images: ', len(os.listdir(train_img_folder)))
```

```python
df = pd.read_csv(bbox_csv)
```

```python
samples = df.sample(4)

for row in samples.values:
    file_path = os.path.join(train_img_folder, row[0])
    img = cv2.imread(file_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2.rectangle(img, (int(row[1]), int(row[2])), (int(row[3]), int(row[4])), (255, 0, 0), 3)
    plt.imshow(img)
    plt.show()
```
